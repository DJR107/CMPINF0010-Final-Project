```python
import pandas as pd

crime = pd.read_csv("PittsburghCrimeReports.csv")
crime.dropna(how='any', subset=['INCIDENTNEIGHBORHOOD'], inplace=True)
    
ncr =  {}
for report in crime.index:
    if crime.loc[report, "INCIDENTNEIGHBORHOOD"] in ncr:
        ncr[crime.loc[report, "INCIDENTNEIGHBORHOOD"]] += 1
    else:
        ncr[crime.loc[report, "INCIDENTNEIGHBORHOOD"]] = 1
        
import collections
ncr = collections.OrderedDict(sorted(ncr.items()))

del ncr["Mount Oliver"]
del ncr["Mount Washington"]
del ncr["Mt. Oliver Boro"]
del ncr["Outside City"]
del ncr["Outside County"]
del ncr["Outside State"]

population = pd.read_csv("population-density.csv")

pop = {}
for town in population.index:
    z = population.loc[town, "Pop. 2010"]
    z = z.replace(',', '')
    pop[population.loc[town, "Neighborhood"]] = z
    
del pop["Chartiers City"]
del pop["Mount Washington"]
del pop["Mt. Oliver"]
    
pop["Central North Side"] = pop["Central Northside"]
del pop["Central Northside"]

pop = collections.OrderedDict(sorted(pop.items()))
    
# print("      Crimes with length " + str(len(ncr)))
    
# for x in ncr:
#     print(x + ": " + str(ncr[x]))
    
# print("      Populations with length " + str(len(pop)))

# for x in pop:
#     print(x + ": " + str(pop[x]))
    
percent = {}
for x in ncr:
    y = int(ncr[x]) / int(pop[x])
    y = round(y, 3)
    percent[x] = (y)
    
del percent["Chateau"]
# del percent["South Shore"]
# del percent["North Shore"]    

import matplotlib.pyplot as plt

plt.figure(figsize=(20, 3))
plt.bar(range(len(percent)), percent.values(), align='center', width=0.3)
plt.xticks(range(len(percent)), percent.keys(), rotation='vertical')
plt.title("Crimes per person in the Neighborhoods of Pittsburgh")
plt.ylabel("Crimes per person")

import operator
percent = sorted(percent.items(), key=lambda kv: kv[1])
percent = collections.OrderedDict(percent)

for x in percent:
    print(x + ": " + str(percent[x]))

# plt.show()

# plt.figure(figsize=(3, 10))
# plt.barh(range(len(first)), list(first.values()), align='center', height=0.3)
# plt.yticks(range(len(first)), list(first.keys()))
# plt.title("Crimes per person in the Neighborhoods of Pittsburgh Part 1")
# plt.xlabel("Crimes per person")

# plt.show()

# plt.figure(figsize=(3, 10))
# plt.barh(range(len(second)), list(second.values()), align='center', height=0.3)
# plt.yticks(range(len(second)), list(second.keys()))
# plt.title("Crimes per person in the Neighborhoods of Pittsburgh Part 2")
# plt.xlabel("Crimes per person")

# plt.show()

# codes = {}
# for num in crime.index:
#     offense = crime.loc[num, "OFFENSES"]
#     sep = '('
#     offense = offense.split(sep, 1)[0]
#     sep1 = ' '
#     offense = offense.split(sep1, 1)[0]
#     sep2 = '.'
#     offense = offense.split(sep2, 1)[0]
    
#     if offense in codes:
#         codes[offense] += 1
#     else:
#         codes[offense] = 1
        
# for x in codes:
#     print(x + ": " + str(codes[x]))
```
