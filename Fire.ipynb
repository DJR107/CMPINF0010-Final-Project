```python
import pandas as pd
%matplotlib inline
import numpy as np
```

```python
fire = pd.read_csv("FireIncidents.csv",index_col="neighborhood", parse_dates=True)
fire.groupby("neighborhood").count()
```

```python
fireAmount = fire.groupby("neighborhood").count()
fireAmount['alarm_time'].plot(kind="bar", width = .5, figsize=(25,10), position=1)
```

```python
def filter_less_than(x,threshold):
    if len(x)>threshold:
        return True
    else:
        return False
manyFire = fire.groupby("fire_zone").filter(filter_less_than, threshold=10)
manyFire.groupby("fire_zone").count()
```
